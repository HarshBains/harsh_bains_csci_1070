# QUESTION 1

Machine learning's primary use cases involve large, unwieldy sets of data that would prove to be far too complex for a programmer to manually preprocess and draw conclusions from. In these cases, machine learning can help in pattern recognition, making predictions (through algorithms such as KNN), visualization, and general dimensonality reduction; more hands-on machine learning models which are generalized as supervised learning, on the other hand, can help in classifying data for manual processing (because the programmer defines the parameters for the classification), as well as demonstrating regression, as illustrated by the class notes. 

Of course, keeping these use cases in mind, there are also times where machine learning is not the most optimal solution for the problem at hand. For instance, if we have concise, quantifiable data that we need to perform a straightforward, deterministic operation on, machine learning needlessly complicates the situation, making traditional functionality a much more prospective pursuit.

Ultimately, then, carefully considering the scope of the problem we're dealing with is vital in determining whether machine learning or conventional programming is the optimal solution. Generally speaking, if we need to perform a calculation that is arithmetic or linear in nature, or if we begin from the onset with clear correlations and a specific outcome, then traditional programming methods will be more suitable to our goal; however, if the problem is multi-layered, with hidden correlations and strutures, then machine learning can help reduce the issue's complexity, and aid in performing more involved operations that elicit a concealed set of features.
